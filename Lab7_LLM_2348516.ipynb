```python
#Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/IMDB Dataset.csv')

df.head()
```

```python


from transformers import BertTokenizer

# Load the BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Check if 'text' column exists, if not, replace 'text' with the correct column name
if 'review' in df.columns:
    # Tokenize the text data
    tokens = df['review'].apply(lambda x: tokenizer.tokenize(x))

    # Encode the tokens
    encoded_data = tokens.apply(lambda x: tokenizer.encode_plus(
        x,
        add_special_tokens=True,
        max_length=128,  # You can adjust this based on your data
        padding='max_length',
        truncation=True,
        return_tensors='pt'
    ))
else:
    print("Column 'text' not found in DataFrame. Please check your CSV file or column name.")
```
